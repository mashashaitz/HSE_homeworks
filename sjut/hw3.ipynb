```python
from wiki_ru_wordnet import WikiWordnet
wikiwordnet = WikiWordnet()
```

```python
synsets = wikiwordnet.get_synsets('замок')
```

```python
lemmas_like = []
for synset in synsets[0:]:
    words = []
    for hypernym in wikiwordnet.get_hypernyms(synset):
        for w in hypernym.get_words():
            words.append(w.lemma())
    for hyponym in wikiwordnet.get_hyponyms(synset):
        for w in hyponym.get_words():
            words.append(w.lemma())
    for w in synset.get_words():
        words.append(w.lemma())
    if len(words) == 1: 
        synsets.remove(synset)
    else:
        lemmas_like.append(words)
print(lemmas_like)
```

```python
for synset in synsets[0:]:
    print('\n')
    print({w.definition() for w in synset.get_words()})
```

```python
scores = []
for n in neighbours:
    t = []
    for ls in lemmas_like:
        tnew = 0
        for wn in n:
            if wn[0] in ls:
                tnew += 1
        t.append(tnew)
    scores.append(t)
print(scores)
```

```python
for n, meaning in enumerate(scores):
    for n2, m in enumerate(meaning):
        if m > 0:
            print(str(n), ' - ', {w.definition() for w in synsets[n2].get_words()})
```
